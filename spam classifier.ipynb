# Data cleaning and preprocessing 

```python
#required library for datacleaning and preprocessing
import re 
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemma = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['messages'][i])
    review = review.lower()
    review = review.split()
    review = [lemma.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = " ".join(review)
    corpus.append(review)
```

```python
###Creating bag of words
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
y = pd.get_dummies(df['labels'])
y
```

```python
y.drop(['ham'],axis=1,inplace=True)
```

```python
##train test split 
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train = train_test_split(x,y,test_size=0.20,random_state =0)
```

```python
#training the model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
confusion_matrix(y_test,y_test)
```

```python
accuracy_score(y_test,y_pred)
```
