# Importing required libraries 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix , classification_report
from sklearn.model_selection import train_test_split
```

# Reading file from local drive

```python
df = pd.read_csv("C:\\Users\\sumeera\\Desktop\\data science\\dataset\\train1.csv")
df.head()
```

```python
target= df['label']
```

```python
x= df.drop(['label'],axis=1)
```

# Converting the data into Dataframe

```python
input_variable = pd.DataFrame(x)
```

```python
input_variable
```

# Beforing doing PCA we need to scale data.so, importing standardscaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
scale_t = scale.fit_transform(input_variable)
```

# Performing PCA

```python
pca = PCA(0.98)
pca_t = pca.fit_transform(scale_t)
```

```python
###eigen values and eigen vector
pca_t
```

# Visualizating PCA

```python
plt.scatter(pca_t[:,0],pca_t[:,1],cmap='prism',c=target)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
inpp = scale.fit_transform(pca_t)
```

```python
inp_variable = pd.DataFrame(inpp)
```

```python
inp_variable
```


# Concating the input and target variables

```python
DFF = pd.concat([target,input_variable],axis=1)
```

# Performing KNN on top of Pca since KNN works Well with High Dimension

```python
b = DFF['label']
a=DFF.drop(['label'],axis=1)
```

# Splitting the data into train anad test

```python
x_test,x_train,y_test,y_train = train_test_split(a,b,test_size =0.2,random_state=123)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train , y_train)
```

```python
y_pred = knn.predict(x_test)
```

# Evaluating the metrics

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
